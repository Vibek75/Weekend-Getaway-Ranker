```python
source_city = "Kolkata"
source = df[df['City'] == source_city].iloc[0]
src_lat, src_lng = source['Latitude'], source['Longitude']
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))
```

```python
df['Distance_km'] = df.apply(
    lambda row: haversine(src_lat, src_lng, row['Latitude'], row['Longitude']),
    axis=1
)
```

```python
weekend_df = df[df['Distance_km'] <= 500].copy()
```

```python
weekend_df['airport_score'] = weekend_df[
    'Airport with 50km Radius'
].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
weekend_df['rating_norm'] = (
    weekend_df['Google review rating'] /
    weekend_df['Google review rating'].max()
)

weekend_df['popularity_norm'] = (
    weekend_df['Number of google review in lakhs'] /
    weekend_df['Number of google review in lakhs'].max()
)

weekend_df['distance_norm'] = 1 - (
    weekend_df['Distance_km'] /
    weekend_df['Distance_km'].max()
)
weekend_df['time_norm'] = (
    1 - (
        weekend_df['time needed to visit in hrs'] /
        weekend_df['time needed to visit in hrs'].max()
    )
)
weekend_df['fee_norm'] = (
    1 - (
        weekend_df['Entrance Fee in INR'] /
        weekend_df['Entrance Fee in INR'].max()
    )
)
```

```python
weekend_df['final_score'] = (
    0.25 * weekend_df['rating_norm'] +
    0.25 * weekend_df['popularity_norm'] +
    0.25 * weekend_df['distance_norm'] +
    0.15 * weekend_df['time_norm'] +
    0.10 * weekend_df['airport_score'] +
    0.05 * weekend_df['fee_norm']
)
```

```python
top_weekend_places = (
    weekend_df
    .sort_values('final_score', ascending=False)
    .head(10)
)

top_weekend_places[
    ['City', 'Name', 'Distance_km', 'Google review rating',
     'Number of google review in lakhs', 'final_score']
]
```
